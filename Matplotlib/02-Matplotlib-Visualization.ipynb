```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
firstNumpyArray = np.linspace(0, 10, 20)
```

```python
secondNumpyArray = firstNumpyArray ** 2
```

```python
(myFigure, myAxis) = plt.subplots()

myAxis.plot(firstNumpyArray, secondNumpyArray, color = "#800080", alpha = 0.5)
myAxis.plot(secondNumpyArray, firstNumpyArray, color = "#008080")
```

```python
(newFigure, newAxis) = plt.subplots()
newAxis.plot(firstNumpyArray, firstNumpyArray + 2, color = "blue", linewidth = 8.0)
newAxis.plot(firstNumpyArray, firstNumpyArray + 3, color = "yellow")
newAxis.plot(firstNumpyArray, firstNumpyArray + 4, color = "black", linestyle = ":")
newAxis.plot(firstNumpyArray, firstNumpyArray + 5, color = "red", linestyle = "-.")
newAxis.plot(firstNumpyArray, firstNumpyArray + 6, color = "pink", linestyle = "--")
newAxis.plot(firstNumpyArray, firstNumpyArray + 7, color = "black", linestyle = "--", marker = "o", markersize = 4, markerfacecolor = "white")
```

## Scatter

```python
plt.scatter(firstNumpyArray, secondNumpyArray)
```
