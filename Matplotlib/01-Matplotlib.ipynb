```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
ages = [10,20,30,40,40,20,10,20,10,70]
```

```python
weight = [20,40,50,60,50,70,90,90,50,10]
```

```python
numpyAgeList = np.array(ages)
numpyWeightList = np.array(weight)
```

```python
plt.plot(numpyAgeList, numpyWeightList, "g")
plt.xlabel("Ages")
plt.ylabel("Weight")
plt.title("graph of weight change with age")
```

```python
plt.plot(numpyWeightList, numpyAgeList, "b")
plt.xlabel("Weight")
plt.ylabel("Ages")
plt.title("graph of weight change with age")
```

```python
firstNumpyArray = np.linspace(0,10,20)
```

```python
secondNumpyArray = firstNumpyArray ** 3
```

```python
plt.plot(firstNumpyArray, secondNumpyArray, "r--")
```

```python
plt.plot(firstNumpyArray, secondNumpyArray, "r*-")
```

```python
plt.subplot(1, 2, 1)
plt.plot(firstNumpyArray, secondNumpyArray, "b--")
plt.subplot(1, 2, 2)
plt.plot(secondNumpyArray, firstNumpyArray, "g*-")
```

```python
myFigure = plt.figure()
figureAxes = myFigure.add_axes([0.2, 0.2, 0.9, 0.9])
figureAxes.plot(firstNumpyArray, secondNumpyArray, "g")
figureAxes.set_xlabel("X")
figureAxes.set_ylabel("Y")
figureAxes.set_title("Graphic")
```

```python
secondFigure = plt.figure()
firstAxis = secondFigure.add_axes([0.1, 0.1, 0.9, 0.9])
secondAxis = secondFigure.add_axes([0.3, 0.4, 0.3, 0.3])

firstAxis.plot(firstNumpyArray, secondNumpyArray, "g")
firstAxis.set_xlabel("First X")
firstAxis.set_ylabel("First Y")
firstAxis.set_title("First Graphic")

secondAxis.plot(secondNumpyArray, secondNumpyArray, "b")
secondAxis.set_xlabel("Second X")
secondAxis.set_ylabel("Second Y")
secondAxis.set_title("Second Graphic")
```

```python
(myFigure, myAxis) = plt.subplots(nrows = 1, ncols = 2)
for axis in myAxis:
    axis.plot(firstNumpyArray, secondNumpyArray, "g")
    axis.set_xlabel("X")
plt.tight_layout()
```

```python
newFigure = plt.figure(dpi= 100)
## dpi 100
## figsize 6,6

newAxis = newFigure.add_axes([0.1, 0.1, 0.9, 0.9])

newAxis.plot(firstNumpyArray, firstNumpyArray ** 2, label = "Numpy Array **2")
newAxis.plot(firstNumpyArray, firstNumpyArray ** 3, label = "Numpy Array **3")
newAxis.legend(loc = 0)
```
